```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("cleaned_BA.csv")
df.head(3)
```

```python
positive=df[df["Rating"]>=7]["Rating"].count()
neutral=df[(df.Rating < 7) & (df.Rating > 4)]["Rating"].count()
negative=df[(df.Rating<=3)]["Rating"].count()
```

```python
plt.figure(figsize=(8,6))

plt.style.use('ggplot')
label=["Positive","Negative","Neutral"]
plt.title("Overall Rating",fontsize=12)

plt.pie([positive,negative,neutral],labels=label,pctdistance=0.8,autopct='%.2f %%',explode=[0,0,0.1])
plt.show()
```

```python
plt.figure(figsize=(8,6))

ratings = df.Rating.to_list()
tips=[0,100,200,300,400,500,600]


plt.yticks([x for x in range(12)])
plt.xticks(tips)
plt.ylabel("ratings")
plt.xlabel("Customer count")
plt.title("Review Analysis",fontsize='13')

plt.hist(ratings,bins=10,orientation='horizontal',color='red',edgecolor='green')
plt.show()


```

```python
r=df.Review.to_list()
review=[]
for i in range(100):
    review.extend(r[i].split())

```

```python


from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)


def show_wordcloud(data):
    wordcloud = WordCloud(
        background_color='white',
        # stopwords=stopwords,
        max_words=100,
        max_font_size=30,
        scale=3,
        random_state=1
        )

    wordcloud=wordcloud.generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')

    plt.imshow(wordcloud)
    plt.show()
    return wordcloud
text=' '.join(review)
cloud=show_wordcloud(text)
print(cloud)
cloud.to_file("wordcloud.png")

```
